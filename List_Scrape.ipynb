```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import numpy as np
import re
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_film_remakes_(A%E2%80%93M)')
soup= BS(response.text)
```

```python
tables=soup.find_all('tr')
tables=[x.text.strip() for x in tables]
tables=[x.split('\n\n') for x in tables if x!='Original\nRemake(s)\nNotes']
tables

```

```python
originals=[]
reems=[]
notes=[]

for x in tables:
    originals.append(x[0])
    reems.append(x[1])
    try:
        notes.append(x[2])
    except:
        notes.append('No Notes')
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_film_remakes_(N%E2%80%93Z)#O')
soup2= BS(response.text)

tables2=soup2.find_all('tr')
tables2=[x.text.strip() for x in tables2]
tables2=[x.split('\n\n') for x in tables2 if x!='Original\nRemake(s)\nNotes']
tables2

for x in tables2:
    originals.append(x[0])
    reems.append(x[1])
    try:
        notes.append(x[2])
    except:
        notes.append('No Notes')
```

```python
split_reems=[x.split(')') for x in reems]
split_reems
```

```python
twos=[]
threes=[]
fours=[]
fives=[]
six_or_more=[]


for x in split_reems:
    if x!='':
        twos.append(x[0])
        try:
            threes.append(x[1])
        except:
            threes.append('No 3rd remake')
        try:
            fours.append(x[2])
        except:
            fours.append('No 4th remake')
        try:
            fives.append(x[3])
        except:
            fives.append('No 5th remake')
        six_or_more.append(x[4:])
    else: continue
    
    
```

```python
twos
```

```python
original_movie=[]
original_year=[]
reem_movie=[]
reem_year=[]
three_movie=[]
three_year=[]
four_movie=[]
four_year=[]
five_movie=[]
five_year=[]


split_originals=[x.split(' (') for x in originals]
for x in split_originals:
    original_movie.append(x[0])
    original_year.append(x[1].strip(')'))
    
new_split_reems=[x.split(' (') for x in twos]
for x in new_split_reems:
    reem_movie.append(x[0])
    reem_year.append(x[1].strip(')'))        

split_threes=[x.split(' (') for x in threes]
for x in split_threes:
    if x != '':
        if x[0]!='':
            try:
                three_movie.append(x[0])
                three_year.append(x[1].strip(')'))
            except:
                continue
three_movie=[x.strip() for x in three_movie if x!='' if x!='No 3rd remake' if x!='.']

split_fours=[x.split(' (') for x in fours]
for x in split_fours:
    if x != '':
        if x[0] !='':
            try:
                four_movie.append(x[0])
                four_year.append(x[1].strip(')'))
            except:
                continue
four_movie=[x.strip() for x in four_movie if x!='' if  x!= 'No 4th remake']


split_five=[x.split(' (') for x in fives]
for x in split_five:
    if x != '':
        if x[0] != '':
            try:
                five_movie.append(x[0])
                five_year.append(x[1].strip(')'))
            except:
                continue
five_movie=[x.strip() for x in five_movie if x!='' if  x!= 'No 5th remake']
```

```python
original_df=pd.DataFrame(original_movie)
original_df['year']=original_year
original_df['title']=original_df[0]
original_df=original_df.drop(columns=0)
original_df
```
